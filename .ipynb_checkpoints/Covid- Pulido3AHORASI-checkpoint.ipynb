```python
import requests
from bs4 import BeautifulSoup
import timeit
import re
import pandas as pd
import numpy as np
```

```python
url = 'https://www.worldometers.info/coronavirus/'
```

```python
%%time
def makesoup(url):
    response = requests.get(url)
    soup= BeautifulSoup(response.content, "html.parser")
    return soup#.find_all('td')
soup = makesoup(url)
```

```python
def get_data(soup):
    return soup.find_all('td')
```
