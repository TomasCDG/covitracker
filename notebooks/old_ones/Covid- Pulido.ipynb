```python
import requests
from bs4 import BeautifulSoup
import timeit
import re
import pandas as pd
import numpy as np
```

```python
url = 'https://www.worldometers.info/coronavirus/'
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# You now can query the `soup` object!
soup.title.string
```

```python
td = soup.find_all('td')
td
```

# get the name of the countries!

```python
%%time
names = soup.find_all('td', style="font-weight: bold; font-size:15px; text-align:left;" )
```

```python
names
```

```python
countries = [i.string for i in names]
```

-----------------------

# get the indices of each nation in the list

```python
for i in enumerate(td):
    print(i)
```

```python
#here we find the index!
```

```python
## %%time

countrylist = countries[0:4]
indexes  = {}
for i in countrylist:
    indexes[i] = 'Notfound'

for i in countrylist:
    j=0
    while indexes[i] != 'NotFound' and j <= len(td):
        if td[j].string == i:
            indexes[i] = j
            break
        else:
            j+=1
```

```python
indexes
```

## do the same for the world!

```python
%%time
def world_stats(x):
    j=0
    while x[j].string != 'World':
        j+=1
    else:
        return j

world_index = world_stats(td)
```
